```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline 
# to plot graphs in the same file
```

```python
# homeprices.csv must be in the same directory as that of the jupyter notebook or code file
# below dataset shows homeprices according to the area 
df = pd.read_csv("homeprices.csv") 
df
```

```python
# labelling the graph
plt.xlabel('area')
plt.ylabel('price(US$)')

#plotting the data
plt.scatter(df.area, df.price, color = 'red')
```

```python
x_values = np.array(df['area']).reshape(-1, 1) # creating numpy array 
y_values = np.array(df.price)  # creating numpy array 

reg = linear_model.LinearRegression() # create Linear regression model or object
reg.fit(x_values, y_values) # put / fit data into the model 
```

```python
# labelling the graph
plt.xlabel('area')
plt.ylabel('price(US$)')

#plotting the given values
plt.scatter(df.area, df.price, color = 'red')

#plotting the regression line by model
plt.plot(x_values, reg.predict(x_values), color = 'blue')
```

```python
reg.predict([[3300]]) # predicting some values on random data
# above code shows prediction of price of home with area = 3300 which
# comes out to be 628715.75$
```
