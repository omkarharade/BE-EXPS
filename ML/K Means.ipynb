```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
y_predicted =  km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Age, df1['Income($)'], color = 'green', label='Income($)')
plt.scatter(df2.Age, df2['Income($)'], color = 'red', label='Income($)')
plt.scatter(df3.Age, df3['Income($)'], color = 'black', label='Income($)')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
# we will scale the data :  that is to put data in range from 0 to 1 using MinMaxScaler
scaler = MinMaxScaler()


# using numpy array to scale income data

income_array = np.array(df[['Income($)']])

scaler.fit(income_array)
income_array = scaler.transform(income_array)
df['Income($)'] = income_array

#without using numpy array to scale age data 

scaler.fit(df[['Age']]) 
df[['Age']] = scaler.transform(df[['Age']])

# any method can work trom these two
```

```python
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
# seperating the data points in three sub data to specify with different color

df1 = df[df.cluster == 0] # df1 contains the data whose predicted cluster came to be 0
df2 = df[df.cluster == 1] # df2 contains the data whose predicted cluster came to be 1
df3 = df[df.cluster == 2] # df3 contains the data whose predicted cluster came to be 2

plt.scatter(df1.Age, df1['Income($)'], color = 'green', label='Income($)')
plt.scatter(df2.Age, df2['Income($)'], color = 'red', label='Income($)')
plt.scatter(df3.Age, df3['Income($)'], color = 'black', label='Income($)')

# DEFINING LABEL FOR GRAPH AXES
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```
