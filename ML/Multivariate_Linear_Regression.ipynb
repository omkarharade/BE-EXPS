```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices2.csv")
df
```

```python
multi_x_values = np.array(df[['area', 'bedrooms', 'age']])
y_values = df.price

reg = linear_model.LinearRegression()
reg.fit(multi_x_values, y_values)
```

```python
# labelling the graph
plt.xlabel('area')
plt.ylabel('price(US$)')

#plotting the data
plt.scatter(df.area, df.price, color = 'red')
```

```python
# labelling the graph
plt.xlabel('bedrooms')
plt.ylabel('price(US$)')

#plotting the data
plt.scatter(df.bedrooms, df.price, color = 'blue')
```

```python
# labelling the graph
plt.xlabel('age')
plt.ylabel('price(US$)')

#plotting the data
plt.scatter(df.age, df.price, color = 'green')
```

```python
# printing the coefficients of the multi regression
reg.coef_
```

```python
#printing the intercept
reg.intercept_
```

```python
# calculating the price for room with 3000 sq ft area, 3 bedrooms, 40 yr old
home_area = 3000
bedrooms = 3
age = 40

predicted_price = reg.predict([[home_area, bedrooms, age]])
predicted_price
```
